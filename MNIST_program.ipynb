```python
import numpy as np
import scipy.special as ss
```

```python
'''以下是神经网络的建立、训练以及测试请求的三个函数的定义'''
```

```python
class neuralnetwork:

    def __init__(self,inputnodes,hiddennodes,outputnodes,learningrate):

        #设置以下三层网络节点数量的变量
        self.inodes = inputnodes
        self.hnodes = hiddennodes
        self.onodes = outputnodes

        #定义权重矩阵以连接不同的网络层
        #wih: weight matrix of the inputnodes and hiddennodes
        self.wih = np.random.normal(0.0,pow(self.hnodes,-0.5),(self.hnodes,self.inodes))#(期望值，方差，矩阵规模)
        self.who = np.random.normal(0.0,pow(self.onodes,-0.5),(self.onodes,self.hnodes))#(期望值，方差，矩阵规模)

        #定义学习率
        self.lr = learningrate

        #定义激活函数（抑制）
        self.actf = lambda x : ss.expit(x)

        pass
        
    def train(self,inputs_list,targets_list):
        
        #转化数组为二维数组并进行转置
        self.inputs = np.array(inputs_list,ndmin= 2).T
        self.targets = np.array(targets_list,ndmin= 2).T

        #计算隐藏层信息和隐藏层信息复处理
        self.hidden_inputs = np.dot(self.wih,self.inputs)
        self.hidden_outputs = self.actf(self.hidden_inputs)

        #计算返回层信息和隐藏层信息复处理
        self.final_inputs = np.dot(self.who,self.hidden_outputs)
        self.final_outputs = self.actf(self.final_inputs)

        #计算误差
        self.outputs_errors = self.targets - self.final_outputs
        self.hidden_errors = np.dot(self.who.T,self.outputs_errors)

        #更新权重矩阵

        self.who += self.lr * np.dot((self.outputs_errors * self.final_outputs * (1.0 - self.final_outputs)),np.transpose(self.hidden_outputs))
        self.wih += self.lr * np.dot((self.hidden_errors * self.hidden_outputs * (1.0 - self.hidden_outputs)),np.transpose(self.inputs))
        

        pass

    def query(self,inputs_list):
        
        #转化数组为二维数组并进行转置
        self.inputs = np.array(inputs_list,ndmin= 2).T

        #计算隐藏层信息和隐藏层信息复处理
        self.hidden_inputs = np.dot(self.wih,self.inputs)
        self.hidden_outputs = self.actf(self.hidden_inputs)

        #计算返回层信息和隐藏层信息复处理
        self.final_inputs = np.dot(self.who,self.hidden_outputs)
        self.final_outputs = self.actf(self.final_inputs)

        return self.final_outputs
```

```python
if __name__ == "__main__":

    #神经网络节点设置
    inputnodes= 784
    hiddennodes= 350
    outputnodes= 10
    learningrate = 0.15
    
    #一号神经网络测试
    test1 = neuralnetwork(inputnodes,hiddennodes,outputnodes,learningrate)
    
    #读取训练数据
    training_data_file = open("train.csv","r")
    training_data_list = training_data_file.readlines()
    training_data_file.close()

    #读取测试数据
    test_data_file = open("train_test.csv","r")
    test_data_list = test_data_file.readlines()
    test_data_file.close()

    #训练神经网络
    era = 3         #世代设置

    for i in range(era):

        #扫描所有数据,调用训练

        print(i) 
        for record in training_data_list:
            #（创建所有值）
            all_value = record.split(',')
            #输入数据处理
            inputs = (np.asfarray(all_value[1:])/255.0*0.99) + 0.01
            #设置理想输出标准
            targets = np.zeros(outputnodes) + 0.01
            targets[int(all_value[0])] = 0.99
            test1.train(inputs,targets)
    #测试神经网络

    #测试计次
    correct = 0
    times = 0

    print("test_start")

    for testrecord in test_data_list:
        times+=1
        test_value = testrecord.split(',')
        label_mc = np.argmax(test1.query(np.asfarray(test_value[1:])/255.0*0.99 + 0.01))
        if(int(test_value[0]) == label_mc):
            correct+=1
            
    # print(test1.query(np.asfarray(test_value[1:])/255.0*0.99 + 0.01),label_mc,test_value[0])
            
    #打印测试结果
    print(correct)
    print(times)
    print(correct*1.0/times)
```
